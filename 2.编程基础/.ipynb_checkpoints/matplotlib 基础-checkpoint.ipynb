```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy, color="red")
```

```python
plt.plot(x,siny)
plt.plot(x,cosy, color="red", linestyle="--")
```

```python
plt.plot(x,siny)
plt.plot(x,cosy, color="red", linestyle="--")
# plt.xlim(-5,15)
# plt.ylim(0,1.5)
plt.axis([-1,11,-2,2])
```

```python
plt.plot(x,siny)
plt.plot(x,cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
```

```python
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy, color="red", linestyle="--",label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.title("Hello")
plt.legend() #
```
